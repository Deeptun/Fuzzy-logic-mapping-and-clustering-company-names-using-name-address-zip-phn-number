```python
import pandas as pd
from rapidfuzz import fuzz
from rapidfuzz import process

# Define the function to calculate similarity score
def calculate_similarity(row, weights):
    # Define the weights for each feature
    name_weight, address_weight, state_weight, zip_weight, phone_weight = weights

    # Calculate the similarity score for each feature
    name_score = fuzz.token_sort_ratio(row['name'], row['name']) * name_weight
    address_score = fuzz.token_sort_ratio(row['address'], row['address']) * address_weight
    state_score = fuzz.token_sort_ratio(row['state'], row['state']) * state_weight
    zip_score = fuzz.token_sort_ratio(row['zip'], row['zip']) * zip_weight
    phone_score = fuzz.token_sort_ratio(row['phone'], row['phone']) * phone_weight

    # Calculate weighted average of scores
    weighted_score = sum([name_score, address_score, state_score, zip_score, phone_score]) / sum(weights)

    return weighted_score

# Define the function to find the closest match for each company
def find_match(row, df, weights, threshold):
    # Calculate the similarity scores between the current row and all other rows
    scores = df.apply(lambda x: calculate_similarity(row, weights) if x['name'] != row['name'] else 0, axis=1)
    
    # Get the index of the closest match (if above threshold)
    best_match_index = scores[scores >= threshold].idxmax()
    
    # Return the name of the best match (or None if no match above threshold)
    return df.loc[best_match_index, 'name'] if pd.notna(best_match_index) else None

# Define the function to cluster the companies
def cluster_companies(df, weights, threshold):
    # Create a copy of the dataframe to avoid modifying the original
    df_clustered = df.copy()
    
    # Create a new column to hold the cluster label
    df_clustered['cluster'] = None
    
    # Iterate over each company
    for i, row in df_clustered.iterrows():
        # If the company has not been assigned to a cluster yet
        if pd.isna(row['cluster']):
            # Find the closest match for the company
            match = find_match(row, df_clustered, weights, threshold)
            
            # If a match was found
            if match is not None:
                # Assign the company to the same cluster as the match
                df_clustered.loc[df_clustered['name'] == match, 'cluster'] = row['name']
                df_clustered.loc[df_clustered['name'] == row['name'], 'cluster'] = row['name']
            else:
                # If no match was found, assign the company to a new cluster
                df_clustered.loc[df_clustered['name'] == row['name'], 'cluster'] = row['name']
                
    # Return the clustered dataframe
    return df_clustered

# Define sample data
data = {'name': ['ABC Inc.', 'ABC Co.', 'XYZ LLC', 'XYZ Corp.', 'PQR Limited', 'PQR Inc.', 'LMN Pvt Ltd.', 'LMN Pvt Ltd'],
        'address': ['123 Main St.', '123 Main St', '456 Park Ave.', '456 Park Ave', '789 Market St.', '789 Market St', '10 Circle Rd.', '10 Circle Road'],
        'state': ['CA', 'CA', 'NY', 'NY', 'NJ', 'NJ', 'MA', 'MA'],
        'zip': ['12345', '12345', '67890', '67890', '45678', '45678', '23456', '23456'],
        'phone': ['555-123-4567', '555-123-4567', '555-456-7890', '555-456-7890', '555-789-0123', '555-789-0123', '555-234-5678', '555-234-5678']}

df = pd.DataFrame(data)
# Define weights for each feature
weights = [0.5, 0.2, 0.1, 0.1, 0.1]
# Define dynamic similarity threshold
threshold = 70

df_nw = cluster_companies(df,weights,threshold)
```
